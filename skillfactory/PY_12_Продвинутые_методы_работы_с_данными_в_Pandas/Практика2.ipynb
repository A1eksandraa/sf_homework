```python
import pandas as pd
rating1 = pd.read_csv('data/ratings1.csv', sep=',')
rating2 = pd.read_csv('data/ratings2.csv', sep=',')
movies = pd.read_csv('data/movies.csv', sep=',')
dates_ = pd.read_csv('data/dates.csv')
```

```python
ratings = pd.concat([rating1, rating2])
ratings
```

```python
ratings = pd.concat(
    [rating1, rating2],
    ignore_index=True
)
ratings
```

```python
print('Число строк в таблице ratings:', ratings.shape[0])
print('Число строк в таблице dates:', dates_.shape[0])
print(ratings.shape[0] == dates_.shape[0])
```

```python
display(rating1.tail(1))
display(rating2.head(1))
```

```python
ratings = ratings.drop_duplicates(ignore_index=True)
print('Число строк в таблице ratings:', ratings.shape[0])
```

```python
ratings_dates = pd.concat([ratings, dates_], axis=1)
ratings_dates
```

```python
joined_false = ratings_dates.join(
    movies,
    rsuffix='_right',
    how='left'
)
joined_false
```

```python
joined = ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how='left'
)
joined
```

```python
merged1 = ratings_dates.merge(
    movies,
    on='movieId',
    how='left'
)
merged1
```

```python
print('Число строк в таблице ratings_dates:', ratings_dates.shape[0])
print('Число строк в таблице merged1:', merged1.shape[0])
print(ratings_dates.shape[0] == merged1.shape[0])
```

```python
merged2 = ratings_dates.merge(
    movies,
    on='movieId',
    how='outer'
)
merged2
```

```python
merge_ratings = rating1.merge(rating2, how='outer')
print('Число строк в таблице merge_ratings:', merge_ratings.shape[0])
merge_ratings
```
